```python
import pandas as pd
import numpy as np

path = 'wine.csv'

# DataFrame으로 읽기
df = pd.read_csv(path, index_col=0) #index_col = 0을 통해서 컬럼갯수 제거
df.head() #빠르게 확긴할 수 있는 명령어
```

```python
# data division (y value)
y = df['Alcohol']    
y.value_counts()
y
```

```python
# data division (x value)
x = df.drop('Alcohol', axis=1)
x.head()
```

```python
# sklearn model split funcation
from sklearn.model_selection import train_test_split

# train_test_split(arrarys, test_size, random_state, shuffle, stratify)
# arrarys : 분할시킬 데이터
# test_size : 학습 데이터 셋의 비율 (float, int)
# random_state : 데이터 분할시 셔플이 이루어지는데 이를 위한 시드값(int, randomState)
# shuffle : 셔플 여부 설정
# stratify : 지정한 Data의 비율
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.5, random_state=0)
```

LinearRegression
DecisionTreeRegressor
RandomForestRegressor
KNN

```python
# 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(x_train, y_train)
ly_preds = model.predict(x_test)

print('평균제곱오차', mean_squared_error(ly_preds, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(x_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(x_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=3)
model.fit(x_train,y_train)
ly_preds = model.predict(x_test)

print('평균제곱오차', mean_squared_error(ly_preds, y_test))
```
