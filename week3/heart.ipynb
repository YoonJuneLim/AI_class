```python
# open csv
import pandas as pd
import numpy as np

path = 'heart.csv'

df = pd.read_csv(path)
df.head()
```

```python
#regression
# age를 통해서 해볼것
# y : age target
# x : age 빼고 나머지
df2 = df.drop('target', axis=1) # target은 뺏음 
y = df2['age']
X = df2.drop('age', axis=1)
X.head()
```

```python
# 훈련을 위한 데이터, 훈련을 위한 정답 데이터
# 테스트 데이터, 이후 나온 값
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=12)

print(X_train.shape, y_train.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
# classification
# label - target (심장병이다 아니다 확인)
# y 값이 target
# x 값이 나머지
y = df['target']
X = df.drop('target',axis=1)
X.head()
```

```python
# 훈련을 위한 데이터, 훈련을 위한 정답 데이터
# 테스트 데이터, 이후 나온 값
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=12)

print(X_train.shape, y_train.shape)
print(y_train.shape, y_test.shape)
```

```python
# Logisstic, DecisionTree, RandomForestClasification
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)
lr_y = clf_lr.predict(X_test)

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, y_train)
dt_y = clf_dt.predict(X_test)

rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
clf_y = rf_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
print("\n--- LogisticRegression Classifier ---")
print(accuracy_score(y_test, lr_y))
print(confusion_matrix(y_test, lr_y))

print("\n--- DecisionTreeClassifier Classifier ---")
print(accuracy_score(y_test, dt_y))
print(confusion_matrix(y_test, dt_y))

print("\n--- RandomForestClassifier Classifier ---")
print(accuracy_score(y_test, clf_y))
print(confusion_matrix(y_test, clf_y))
```
