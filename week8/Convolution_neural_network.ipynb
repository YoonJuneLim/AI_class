```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. 이미지 불러오기 (이미지 경로를 본인의 파일명으로 수정하세요)
# 예: 'my_picture.jpg'
image_path = 'fruits.jpg' 
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE) # 흑백으로 읽기

if img is None:
    print("이미지를 찾을 수 없습니다. 경로를 확인해주세요.")
else:
    # 2. 세로선(Vertical Edge)을 찾는 필터 정의
    # 왼쪽은 -1, 오른쪽은 1로 설정하여 '밝기 차이'를 극대화합니다.
    v_kernel = np.array([[-1, 0, 1],
                         [-2, 0, 2],
                         [-1, 0, 1]])

    # 3. OpenCV의 filter2D 함수로 합성곱 수행
    # 엑셀에서 수동으로 하던 계산을 이 함수가 한 번에 처리합니다.
    vertical_edge = cv2.filter2D(img, -1, v_kernel)

    # 4. 결과 시각화
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.title("Original Gray Image")
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Vertical Edge Detection")
    plt.imshow(vertical_edge, cmap='gray')
    plt.axis('off')

    plt.show()
```
