딥러닝 핵심 요약

[모델 코드]
model = Sequential()
model.add(Dense(10, input_shape=(30,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(2, activation='softmax'))
model.compile(Adam(learning_rate=0.04), 'binary_crossentropy', metrics=['accuracy'])

[한 줄 설명]
입력 30개를 받아 은닉층(10->8->6)을 거쳐, 최종 2개 클래스 확률을 출력하는 분류 모델

[input_shape 중요도]
- input_shape=(30,)은 "입력 특성 30개"를 의미
- 첫 층의 기준 차원을 정하므로 모델의 시작점
- 데이터 열 개수와 다르면 학습/예측 자체가 불가능

[출력층 중요도]
- 출력층은 "무엇을 예측하는지"를 결정하는 핵심
- Dense(2)+softmax: 2개 클래스 확률 출력
- 출력층과 손실함수 조합이 맞아야 학습이 정상 동작

[손실함수 매칭 핵심]
- Dense(1)+sigmoid -> binary_crossentropy (이진분류에서 가장 흔함)
- Dense(2)+softmax -> categorical_crossentropy / sparse_categorical_crossentropy 계열
- 현재 코드는 Dense(2)+softmax + binary_crossentropy 조합이라, 실무에서는 손실함수 재검토 권장

[중요한 것 확인하기]
1) 입력 열 개수 == input_shape인지
2) 출력 노드 수 == 클래스 수인지
3) 출력 활성화(softmax/sigmoid)와 손실함수 조합이 맞는지
4) 학습률(0.04)이 너무 크지 않은지 (불안정하면 0.001~0.01부터 재탐색)

[참고]
최신 Keras 표기: Adam(lr=...) 대신 Adam(learning_rate=...)

원핫 인코딩(One-Hot Encoding) 핵심

- 의미: 범주형 값을 0/1 벡터로 바꾸는 방법
- 예시: 색상 {red, blue, green} -> red=[1,0,0], blue=[0,1,0], green=[0,0,1]
- 이유: 모델이 범주를 숫자 크기(순서)로 오해하지 않게 함
- 용도: 레이블을 클래스별로 명확히 구분하기 위해 사용
- 주의: 범주 수가 많으면 차원이 커지므로 임베딩(Embedding) 사용 고려
</div>

추가 내용 (원본 유지)

모델 코드 핵심
- 입력층 기준: input_shape=(30,) -> 입력 특성 30개
- 은닉층: 10 -> 8 -> 6 (ReLU)
- 출력층: Dense(2) + softmax -> 2개 클래스 확률 출력

중요도 체크
1) 입력 열 개수와 input_shape 일치 여부
2) 출력 노드 수와 클래스 수 일치 여부
3) 출력 활성화와 손실함수 조합 일치 여부
4) 학습률이 너무 크지 않은지

경사하강법 핵심
- W(t+1) = W(t) - η * G
- 핵심은 다음 가중치 W(t+1)를 손실이 줄어드는 방향으로 업데이트하는 것

원핫 인코딩 용도
- 레이블을 클래스별로 명확히 구분하기 위해 사용
</div>

<div style="font-size:13px; line-height:1.45;">
기존 5가지 핵심 (복구)

1. Forward propagation
입력 데이터를 통과시켜 예측값을 계산

2. Backpropagation
오차를 뒤로 전파해 기울기 계산

3. Activation function
비선형성을 부여해 복잡한 패턴 학습

4. Loss function
예측 오차를 수치로 계산

5. Optimization (SGD 등)
기울기로 가중치를 업데이트해 손실 감소
</div>
